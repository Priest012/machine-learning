```python
x = customer.loc[:,['Age', 'Income','Family', 'CCAvg',
       'Education', 'Mortgage', 'Securities Account',
       'CD Account', 'Online', 'CreditCard']]
```

```python
x.shape
```

```python
y = customer.iloc[:,9]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

Using logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit = LogisticRegression()
```

```python
logit.fit(x_train,y_train)
```

```python
ypred = logit.predict(x_test)
```

```python
from sklearn import metrics
```

```python
pcplogit = metrics.confusion_matrix(y_test,ypred) # PCPLogit means percent correctly predicted for logit model.
```

```python
pcplogit
```

```python
print("Accuracy:",round(metrics.accuracy_score(y_test,ypred)*100,2),'%')
```

Naive bayes model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
yprednb = model.predict(x_test)
```

```python
pcpnb = metrics.confusion_matrix(y_test,yprednb) # Percent correctly predicted for naive bayes
```

```python
pcpnb
```

```python
print("Accuracy:",round(metrics.accuracy_score(y_test,yprednb)*100,2),'%')
```

As the accuracy of Logistic regression is more than the accuracy of Naive bayes classifier therefore we are going to finalize this model.
